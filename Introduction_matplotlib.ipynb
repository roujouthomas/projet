### Multiples figures

Il est possible de représenter plusieurs figures sur une même fenêtre de visualisation:

```python
x = numpy.linspace(0, 2*numpy.pi, 30)
fig = matplotlib.pyplot.figure()

ax1 = fig.add_subplot(121)
matplotlib.pyplot.xscale("log")
ax2 = fig.add_subplot(122)
matplotlib.pyplot.xscale("log")
y1 = numpy.cos(x)
y2 = numpy.sin(x)
ax1.plot(x, y1, "g.--",linewidth=1)

ax1.set_title("Cosinus")
ax1.set_xlabel("angle en radian")
ax1.set_ylabel("Cosinus(x)")

ax2.plot(x, y2, "m:x")
ax2.set_title("Sinus",color = 'b')
ax2.set_xlabel("angle en radian")
ax2.set_ylabel("Sinus(x)")
```

### Modification des Légendes avec nom dépendant de variables

```python
def gaussienne(x, m):
    y = numpy.exp(-(x - m)**2)
    return y
```

```python
x = numpy.linspace(0, 20, 100)
m1 = 3 
m2 = 6
y1 = gaussienne(x,m1)
y2 = gaussienne(x,m2)
matplotlib.pyplot.plot(x, y1, "g.--",linewidth=1)
matplotlib.pyplot.plot(x, y2, "r.--",linewidth=1)
matplotlib.pyplot.legend(["m = {}" .format(m1) ,"m = {} ".format(m2) ])
```

## Exercice 1: Marche aléatoire 1D
Représentez pour un M, p et N fixés le déplacement de la marche aléatoire 1D que vous avez implémentée lors de la manipulation précédente ainsi qu'une figure représentant les moyennes et écart-types obtenus en fonction de M et les valeurs théoriques attendues. Vous choisirez la représentation qui vous paraît être la plus appropriée (couleur, titre, légendes doivent être telles que la courbe doit se suffise à elle-même).

```python
import numpy as np
import matplotlib.pyplot as plt

def random_walk_1D(p, N):
    steps = np.random.choice([-1, 1], size=N, p=[1-p, p])
    # Somme des pas pour obtenir le déplacement final
    displacement = np.sum(steps)
    return displacement

# Exemple : p = 0.6, N = 100
p = 0.6
N = 100
displacement = random_walk_1D(p, N)
print(f"Déplacement après {N} pas avec p = {p} : {displacement}")

def monte_carlo_walk_1D(p, N, M):
    deplacement = []
    for _ in range(M):
        deplacement.append(random_walk_1D(p, N))
    deplacement = np.array(deplacement)
    
    # Moyenne et écart-type
    moyenne = np.mean(deplacement)
    écart_type = np.std(deplacement)
    return deplacement, moyenne, écart_type

# Exemple : p = 0.6, N = 100, M = 1000
M = 100
deplacement, moyenne, écart_type = monte_carlo_walk_1D(p, N, M)
print(f"Moyenne des déplacements pour M = {M} : {moyenne}")
print(f"Écart-type des déplacements pour M = {M} : {écart_type}")

valeur_théorique_moyenne = N * (2 * p - 1)
valeur_théorique_écart_type = np.sqrt(N * 4 * p * (1 - p))

plt.figure(figsize=(10, 5))
plt.plot(deplacement, '.', alpha=0.5, label="Déplacement simulé")
plt.axhline(valeur_théorique_moyenne, color='r', linestyle='--', label="Moyenne théorique")
plt.axhline(valeur_théorique_moyenne + valeur_théorique_écart_type, color='g', linestyle='--', label="Écart-type théorique (supérieur)")
plt.axhline(valeur_théorique_moyenne - valeur_théorique_écart_type, color='g', linestyle='--', label="Écart-type théorique (inférieur)")
plt.xlabel("Simulation")
plt.ylabel("Déplacement")
plt.legend()
plt.grid()
plt.show()


M_values = np.arange(10, M+1, 10)
moyennes_simulées = []
écarts_types_simulés = []

for m in M_values:
    _, moyenne_m, écart_type_m = monte_carlo_walk_1D(p, N, m)
    moyennes_simulées.append(moyenne_m)
    écarts_types_simulés.append(écart_type_m)


moyennes_simulées = np.array(moyennes_simulées)
écarts_types_simulés = np.array(écarts_types_simulés)

plt.figure(figsize=(10, 5))
plt.plot(M_values, moyennes_simulées, label="Moyenne simulée", color='b')
plt.axhline(valeur_théorique_moyenne, color='r', linestyle='--', label="Moyenne théorique")
plt.plot(M_values, écarts_types_simulés, label="Écart-type simulé", color='g')
plt.axhline(valeur_théorique_écart_type, color='orange', linestyle='--', label="Écart-type théorique")
plt.xlabel("Nombre de simulations (M)")
plt.ylabel("Valeurs")
plt.legend()
plt.grid()
plt.show()
```

## Exercice 3: Fourmi de Langton
Une fourmi se déplace sur une grille. Au départ toutes les cases sont blanches et la fourmi est positionnée au centre de la grille tête vers le haut.
Si la fourmi est sur une case blanche, elle effectue une rotation  d'un quart de tour vers la gauche, si elle est sur une case noire elle effectue une rotation d' un quart de tour vers la droite. La fourmi inverse la couleur de la case sur laquelle elle se trouve puis avance d'une case dans la direction de son orientation. Programmez le déplacement de la fourmi et le représenter. 
Observez le comportement de sa trajectoire lorsque le nombre de dépacement est supérieur à 10 000.

```python
#exercice 2

import numpy as np
import matplotlib.pyplot as plt

def random_walk_2D(p, N):
    L = np.zeros((N, 2))
    for i in range(N):
        a = np.random.rand()  
        if a >= 0.5:  
            b = np.random.rand()
            if p > b:
                L[i][0] = 1  # Droite
            else:
                L[i][0] = -1  # Gauche
        else:  # Mouvement vertical
            b = np.random.rand()
            if p > b:
                L[i][1] = 1  # Haut
            else:
                L[i][1] = -1  # Bas
    
   
    return np.cumsum(L, axis=0)

# Paramètres
p = 0.6
N = 100

# Génération de la marche aléatoire
positions = random_walk_2D(p, N)
x, y = positions[:, 0], positions[:, 1]


plt.figure(figsize=(8, 8))
plt.plot(x, y, marker='o', markersize=3, label="Trajectoire")
plt.title(f"Marche aléatoire 2D ({N} pas, p={p})")
plt.xlabel("Position X")
plt.ylabel("Position Y")

```

```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt


def fourmis(taille, pas):
    grid = np.zeros((taille, taille),)  
    x, y = taille // 2, taille // 2 
    direction = 0  
    dx = [0, 1, 0, -1]
    dy = [1, 0, -1, 0]
    
    trajectoire_x = [x]
    trajectoire_y = [y]
    
    for i in range(pas):
        if grid[x, y] == 0:  
            direction = (direction - 1) % 4 #gauche
        else:  # Case noire
            direction = (direction + 1) % 4 #droite
    
        grid[x, y] = 1 - grid[x, y]
        
        x += dx[direction]
        y += dy[direction]
        

        trajectoire_x.append(x)
        trajectoire_y.append(y)
    
    return trajectoire_x, trajectoire_y

# Paramètres
taille = 201  # Taille de la grille (doit être suffisamment grande)
pas = 12000  # Nombre de déplacements


trajectoire_x, trajectoire_y = fourmis(taille, pas)



plt.plot(trajectoire_x, trajectoire_y, linewidth=0.5, color='blue')
plt.title(f"Fourmi de Langton-{pas} étapes")
plt.xlabel("Position X")
plt.ylabel("Position Y")
plt.grid(alpha=0.3)
plt.axis("equal")
plt.show()
```
