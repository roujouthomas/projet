# Seaborn

**Seaborn** https://seaborn.pydata.org est une librairie python pour la visualisation des données. Cette librairie a été développée pour compléter et étendre **matplotlib**. Elle permet de résoudre les difficultés majeures rencontrées avec  **matplotlib** : les paramètres par défaut ne sont pas les meilleurs choix et travailler avec des data frame n'est pas directement possible avec **matplotlib**, il faut extraire des Series du même type. **Seaborn**  fonctionne avec **Numpy** et **Pandas**. Les graphiques obtenus sont aussi visuellement plus "design" qu'avec **matplotlib** et surtout plus faciles à réaliser car beaucoup de thèmes sont utilisables alors qu'avec **matplotlib**, il faut le configurer soit même.

**Importation des librairies**

```python
import pandas
import numpy
import matplotlib
import seaborn
```

**Création d'une fonction gaussienne**

```python
def gausplot():
    x = numpy.linspace(0, 20, 100)
    for i in range(1, 20):
        matplotlib.pyplot.plot(x, numpy.exp(-(x - i)**2))
```

```python
gausplot()
```

### Représentation Seaborn
Seaborn dispose d'outils de visualisation haut niveau par rapport à **matplotlib**, il est possible d'utiliser les commandes de matplotlib avec le graphisme de **seaborn**. Pour cela il suffit d'utiliser la fonction *set()* 

```python
seaborn.set()
gausplot()
matplotlib.pyplot.legend('ABCDEF', ncol=2, loc='upper left');
```

# Différents types de représentation 
* barres verticales https://seaborn.pydata.org/generated/seaborn.catplot.html
* barres horizontales https://seaborn.pydata.org/generated/seaborn.countplot.html?highlight=countplot#seaborn.countplot
* Points https://seaborn.pydata.org/generated/seaborn.scatterplot.html
* Multi-barres (groupées ou empilées) https://seaborn.pydata.org/generated/seaborn.catplot.html, https://seaborn.pydata.org/generated/seaborn.barplot.html
* Carte de chaleur https://seaborn.pydata.org/generated/seaborn.heatmap.html



## Barres verticales

```python
seaborn.set()
planets = seaborn.load_dataset("planets")
print(planets.head())
g = seaborn.catplot(x = "year", data=planets, aspect=2,
                       kind="count", palette=["m", "g"])
g.set_xticklabels(step=5)
```

## Barres horizontales

```python
g = seaborn.countplot(y = "year", data=planets,  palette="viridis")
```

## Points

```python
p = (planets
     .groupby(['year']).count()
    )
p['number']

seaborn.scatterplot(x = p.index,y = p['number'] )
```

## Barres multiples verticales

```python
g = seaborn.catplot(x = "year", data=planets, aspect=4.0, kind='count',
                       hue='method', order=range(2001, 2015))
g.set_ylabels('Number of Planets Discovered')
```

```python
g = seaborn.catplot(x = "year", hue="method", kind="count",
            palette="pastel", edgecolor=".6", height =10,
            data=planets);
g.set_xlabels('Number of Planets Discovered')
```

## Barres multiples horizontales

```python
g = seaborn.catplot(y="year", hue="method", kind="count",
            palette="pastel", edgecolor=".6",height =10,
            data=planets);
g.set_xlabels('Number of Planets Discovered')
```

```python
p = (planets
     .groupby(['year']).count()
    )
p.plot(kind='barh', stacked=True, width=1, lw=0.1)
```

```python
p.plot(kind='bar', stacked=True, width=1, lw=0.1)
```

## Cartes de chaleur

```python
p = (planets
     .groupby(['year']).count()
    )
p
f, ax = matplotlib.pyplot.subplots(figsize=(9, 6))
seaborn.heatmap(p, annot=True, fmt="d", cmap= 'hot', linewidths=.5, ax=ax)
```

## Bubble chart
https://seaborn.pydata.org/generated/seaborn.relplot.html

```python
seaborn.set()# pour revenir aux paramètres par defaut
g = seaborn.relplot(
    x="horsepower", 
    y="mpg", 
    hue="origin", 
    size="weight",
    sizes=(40, 400), 
    alpha=.5, 
    palette="Spectral",
    height=6, 
    data=mpg)
g.set_axis_labels("Puissance", "Miles par gallon");
```

## Multiples plots sur une même figure

```python
fig = matplotlib.pyplot.figure(figsize=(10,10))
fig.suptitle("Les exoplanètes")
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)
seaborn.heatmap(p, annot=True, fmt="d", cmap= 'hot', linewidths=.5, ax=ax1)
p.plot(kind='barh', stacked=True, width=1, lw=0.1, ax=ax2)
seaborn.countplot(y = "year", data=planets,  palette="viridis",ax=ax3)
seaborn.scatterplot(x= p.index,y = p['number'], ax= ax4 )
```

## Modification de la taille des figures, des fonts, des titres

```python
fig = matplotlib.pyplot.figure(figsize=(50,50))
#seaborn.set(font_scale=5)
ax = seaborn.lineplot(
    x="fare", 
    y="age",
    ci = None,#sans barres d'erreur,
    color='b',
    markers=True,
    linewidth = 5,
    data=titanic)
ax.set(xlim=(0,100))
ax.set_title("Passager du Titanic ",fontsize = 80)
ax.set_xlabel("Tarif", fontsize = 80)
ax.set_ylabel("Age", fontsize = 80)
```

## Sauvegarder la figure
Les formats supportés sont png, pdf, eps, svg, jpg.

```python
g = seaborn.lmplot(
    x="horsepower", 
    y="mpg", 
    hue="origin",height=7, aspect=1.6,
    data=mpg);
matplotlib.pyplot.savefig('image.png',dpi= 'figure'); 
```

```python

disciplines = data.groupby(['Year', 'Discipline']).size().reset_index(name='Medals')
disciplines = disciplines.pivot(index='Year', columns='Discipline', values='Medals')


disciplines = disciplines.fillna(0)


f, ax = plt.subplots(figsize=(12, 8))
seaborn.heatmap(disciplines, annot=True, fmt="g", cmap='hot', linewidths=0.01, ax=ax)
plt.title("Nombre de médailles par année et par discipline")
plt.ylabel("Année")
plt.xlabel("Discipline")
plt.tight_layout()
plt.show()

```

```python
medailles_par_pays = data.groupby(['Year', 'Country']).size().reset_index(name='Medals')
heatmap_data = medailles_par_pays.pivot(index='Year', columns='Country', values='Medals')
heatmap_data = heatmap_data.fillna(0)
plt.figure(figsize=(20, 10))
seaborn.heatmap(heatmap_data, cmap="YlGnBu", linewidths=0.1, linecolor='black', annot=False)
plt.title("Nombre de médailles par pays et par année")
plt.xlabel("Pays")
plt.ylabel("Années")
plt.show()

#le trop gros nombres de medailles de l'URSS et des états-Unis empeche  de bien voir le nombre de médailles par pays
```

```python
medailles = data.groupby(['Year', 'Country']).size().reset_index(name='Medals')


plt.figure(figsize=(12, 6))
seaborn.scatterplot(
    data=medailles,
    x='Year',        
    y='Medals',      
    hue='Country',    
    alpha=0.7        
)

plt.title("Nombre de médailles par pays et par année")
plt.xlabel("Année")
plt.ylabel("Nombre de médailles")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') 
plt.show()

#toujours pas très lisibiles mise à part pour les surperformances
```

```python
medailles = data.groupby(['Year', 'Country']).size().reset_index(name='Medals')
medailles_pivot = medailles.pivot(index='Year', columns='Country', values='Medals').fillna(0)


plt.figure(figsize=(16, 8))
medailles_pivot.plot(kind='bar', stacked=True, figsize=(16, 8), colormap='tab20')

plt.title("Nombre de médailles par pays et par année")
plt.xlabel("Année")
plt.ylabel("Nombre de médailles")
plt.legend(title='Pays', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assurer que toutes les colonnes importantes sont bien formatées
data = data.dropna(subset=['Year', 'Discipline', 'NOC', 'Gender', 'Medal'])

# --- Visualisation 1 : Répartition des médailles par pays (Top 10) ---
top_countries = (
    data['NOC'].value_counts().head(10).reset_index()
)
top_countries.columns = ['Country', 'Medals']

plt.figure(figsize=(10, 6))
sns.barplot(data=top_countries, x='Medals', y='Country', palette='viridis')
plt.title('Top 10 des pays ayant remporté le plus de médailles', fontsize=14)
plt.xlabel('Nombre de médailles', fontsize=12)
plt.ylabel('Pays', fontsize=12)
plt.tight_layout()
plt.show()
```
